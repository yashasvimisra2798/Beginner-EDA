```
import pandas as pd
import numpy as np
import seaborn as sns
```

```
data = pd.read_csv('titanic.csv')
```

```
sns.scatterplot(x="Age", y="Fare", hue = 'Pclass', data=data)
```

```
sns.boxplot(x="Pclass", y="Fare",data= data)
```

```
sns.distplot(data['Pclass'], kde=False)
```

```
sns.catplot(x="Embarked", y="Survived", hue="Sex",col="Pclass", kind = 'bar',data=data, palette = "rainbow")
```

```
sns.catplot(x="Embarked", y="Survived", hue="Sex",col="Pclass", kind = 'violin',data=data, palette = "rainbow")
```

```
sns.violinplot(x='Sex', y='Survived',data=data)
```

```
sns.pairplot(data, hue="Sex")
```
